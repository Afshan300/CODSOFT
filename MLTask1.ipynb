```python
import pandas as pd

train_path = "movie_data/Genre Classification Dataset/train_data.txt"
test_solution_path = "movie_data/Genre Classification Dataset/test_data_solution.txt"

# IMPORTANT: correct separator is ':::'
df_train = pd.read_csv(
    train_path,
    sep=":::",
    engine="python",
    names=["title", "genre", "plot"],
    dtype=str
)

df_test = pd.read_csv(
    test_solution_path,
    sep=":::",
    engine="python",
    names=["title", "genre", "plot"],
    dtype=str
)

print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)
print(df_train.head())
```

```python
# Remove rows with missing plot or genre
df_train = df_train.dropna(subset=["plot", "genre"])
df_test = df_test.dropna(subset=["plot", "genre"])

# Convert to string and strip spaces
df_train["plot"] = df_train["plot"].astype(str).str.strip()
df_train["genre"] = df_train["genre"].astype(str).str.strip()

df_test["plot"] = df_test["plot"].astype(str).str.strip()
df_test["genre"] = df_test["genre"].astype(str).str.strip()

# Remove empty plots (length 0)
df_train = df_train[df_train["plot"].str.len() > 0]
df_test = df_test[df_test["plot"].str.len() > 0]

print("Train shape after cleaning:", df_train.shape)
print("Test shape after cleaning:", df_test.shape)
```

```python
X_train_text = df_train["plot"]
y_train = df_train["genre"]

X_test_text = df_test["plot"]
y_test = df_test["genre"]

print("Train samples:", len(X_train_text))
print("Test samples:", len(X_test_text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=5000,   # top 5000 words
    ngram_range=(1, 1)   # sirf single words
)

print("Fitting TF-IDF on training data...")
X_train_tfidf = vectorizer.fit_transform(X_train_text)

print("Transforming test data...")
X_test_tfidf = vectorizer.transform(X_test_text)

print("TF-IDF train shape:", X_train_tfidf.shape)
print("TF-IDF test shape:", X_test_tfidf.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000, n_jobs=-1)

print("Training model...")
model.fit(X_train_tfidf, y_train)
print("Training done!")
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```
